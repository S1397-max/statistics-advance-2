7. Hypothesis tests can be one-tailed or two-tailed, depending on the research question and expected direction of results.

One-Tailed Test:

Tests for an effect in one specific direction (either greater or smaller).

Used when prior knowledge suggests the effect will move in one direction.

Example: Testing whether a new drug reduces blood pressure. If the test focuses only on reductions, it is a one-tailed test.

Advantage: More statistical power in detecting an effect, since all significance is concentrated in one tail.

Two-Tailed Test:

Tests for an effect in either direction (increase or decrease).

Used when the direction of the effect is unknown.

Example: Testing whether a new teaching method changes student performance (it could improve or worsen scores).

Advantage: More rigorous, as it accounts for changes in both directions.

Mathematically, a one-tailed test focuses on either the upper or lower tail of the normal distribution, whereas a two-tailed test examines both tails, making the criteria for significance stricter.

```python
# prompt: Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
# interpret the results

import numpy as np
from scipy import stats

def perform_z_test(sample_mean, population_mean, population_std_dev, sample_size, alpha=0.05, alternative='two-sided'):
  """
  Performs a Z-test and interprets the results.

  Args:
    sample_mean (float): The mean of the sample.
    population_mean (float): The known mean of the population.
    population_std_dev (float): The known standard deviation of the population.
    sample_size (int): The number of observations in the sample.
    alpha (float, optional): The significance level (alpha). Defaults to 0.05.
    alternative (str, optional): The alternative hypothesis. Can be 'two-sided',
                                 'less', or 'greater'. Defaults to 'two-sided'.

  Returns:
    tuple: A tuple containing the Z-statistic, p-value, and interpretation.
  """
  # Calculate the Z-statistic
  z_statistic = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

  # Calculate the p-value
  if alternative == 'two-sided':
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
  elif alternative == 'less':
    p_value = stats.norm.cdf(z_statistic)
  elif alternative == 'greater':
    p_value = 1 - stats.norm.cdf(z_statistic)
  else:
    return None, None, "Invalid alternative hypothesis. Choose 'two-sided', 'less', or 'greater'."

  # Interpret the results
  interpretation = f"Z-statistic: {z_statistic:.4f}, P-value: {p_value:.4f}\n"
  if p_value < alpha:
    interpretation += f"Since the p-value ({p_value:.4f}) is less than alpha ({alpha}), we reject the null hypothesis."
    if alternative == 'two-sided':
        interpretation += "\nThere is sufficient evidence to suggest that the sample mean is significantly different from the population mean."
    elif alternative == 'less':
        interpretation += "\nThere is sufficient evidence to suggest that the sample mean is significantly less than the population mean."
    elif alternative == 'greater':
        interpretation += "\nThere is sufficient evidence to suggest that the sample mean is significantly greater than the population mean."

  else:
    interpretation += f"Since the p-value ({p_value:.4f}) is greater than or equal to alpha ({alpha}), we fail to reject the null hypothesis."
    interpretation += "\nThere is not enough evidence to suggest that the sample mean is significantly different from the population mean."

  return z_statistic, p_value, interpretation

# Example Usage:
# Assume we want to test if the average height of a sample of 100 students
# is significantly different from the known population average height of 165 cm.
# The population standard deviation is known to be 10 cm.

sample_mean = 168
population_mean = 165
population_std_dev = 10
sample_size = 100
alpha = 0.05

z_stat, p_val, interpretation = perform_z_test(sample_mean, population_mean, population_std_dev, sample_size, alpha)
print(interpretation)

# Example with a one-sided test (is the sample mean significantly greater?)
z_stat_greater, p_val_greater, interpretation_greater = perform_z_test(sample_mean, population_mean, population_std_dev, sample_size, alpha, alternative='greater')
print("\nOne-sided test (greater):")
interpretation_greater
```

```python
# prompt:  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

def visualize_errors(null_mean, alt_mean, std_dev, sample_size, alpha):
    """
    Calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

    Args:
        null_mean (float): The mean under the null hypothesis (H0).
        alt_mean (float): The mean under the alternative hypothesis (H1).
        std_dev (float): The population standard deviation.
        sample_size (int): The number of observations in the sample.
        alpha (float): The significance level (Type 1 error rate).
    """
    # Standard error
    std_error = std_dev / np.sqrt(sample_size)

    # Calculate critical value (Z-score) for Type 1 error
    # For a one-tailed test (assuming H1 is greater than H0)
    # Adjust for two-tailed if necessary based on the test you are simulating
    critical_z = norm.ppf(1 - alpha)
    critical_value_sample_mean = null_mean + critical_z * std_error

    # Probability of Type 2 error (Beta)
    # This is the probability of failing to reject H0 when H1 is true.
    # We calculate the probability of observing a sample mean less than the critical value
    # under the distribution defined by the alternative hypothesis.
    beta = norm.cdf(critical_value_sample_mean, loc=alt_mean, scale=std_error)

    # Power of the test (1 - Beta)
    power = 1 - beta

    # Visualization
    x = np.linspace(min(null_mean, alt_mean) - 4 * std_error, max(null_mean, alt_mean) + 4 * std_error, 1000)
    pdf_null = norm.pdf(x, loc=null_mean, scale=std_error)
    pdf_alt = norm.pdf(x, loc=alt_mean, scale=std_error)

    plt.figure(figsize=(12, 7))
    plt.plot(x, pdf_null, label=f'Null Hypothesis (H0): Œº = {null_mean}')
    plt.plot(x, pdf_alt, label=f'Alternative Hypothesis (H1): Œº = {alt_mean}', linestyle='dashed')

    # Shade Type 1 error region (reject H0 when H0 is true)
    x_type1 = x[x >= critical_value_sample_mean]
    pdf_null_type1 = pdf_null[x >= critical_value_sample_mean]
    plt.fill_between(x_type1, pdf_null_type1, color='red', alpha=0.5, label=f'Type 1 Error (Œ± = {alpha:.2f})')

    # Shade Type 2 error region (fail to reject H0 when H1 is true)
    x_type2 = x[x <= critical_value_sample_mean]
    pdf_alt_type2 = pdf_alt[x <= critical_value_sample_mean]
    plt.fill_between(x_type2, pdf_alt_type2, color='orange', alpha=0.5, label=f'Type 2 Error (Œ≤ = {beta:.4f})')

    # Add critical value line
    plt.axvline(critical_value_sample_mean, color='black', linestyle='dotted', linewidth=2, label=f'Critical Value ({critical_value_sample_mean:.4f})')

    plt.title('Visualization of Type 1 and Type 2 Errors')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Calculated Critical Value (sample mean): {critical_value_sample_mean:.4f}")
    print(f"Probability of Type 1 Error (Œ±): {alpha:.4f}")
    print(f"Probability of Type 2 Error (Œ≤): {beta:.4f}")
    print(f"Power of the test (1-Œ≤): {power:.4f}")

# Example Usage:
# Assume H0: mean = 100, H1: mean = 105, std_dev = 15, sample_size = 30, alpha = 0.05
visualize_errors(null_mean=100, alt_mean=105, std_dev=15, sample_size=30, alpha=0.05)
```

```python
# prompt: Write a Python program to perform an independent T-test and interpret the results

def perform_independent_ttest(group1_data, group2_data, alpha=0.05, alternative='two-sided', equal_var=True):
  """
  Performs an independent samples T-test and interprets the results.

  Args:
    group1_data (list or np.array): Data for the first group.
    group2_data (list or np.array): Data for the second group.
    alpha (float, optional): The significance level (alpha). Defaults to 0.05.
    alternative (str, optional): The alternative hypothesis. Can be 'two-sided',
                                 'less', or 'greater'. Defaults to 'two-sided'.
    equal_var (bool, optional): Assumes equal population variances if True. Defaults to True.

  Returns:
    tuple: A tuple containing the T-statistic, p-value, and interpretation.
  """
  # Perform independent samples t-test
  t_statistic, p_value = stats.ttest_ind(group1_data, group2_data, alternative=alternative, equal_var=equal_var)

  # Interpret the results
  interpretation = f"T-statistic: {t_statistic:.4f}, P-value: {p_value:.4f}\n"

  if p_value < alpha:
    interpretation += f"Since the p-value ({p_value:.4f}) is less than alpha ({alpha}), we reject the null hypothesis."
    if alternative == 'two-sided':
        interpretation += "\nThere is sufficient evidence to suggest a significant difference between the means of the two groups."
    elif alternative == 'less':
        interpretation += "\nThere is sufficient evidence to suggest that the mean of Group 1 is significantly less than the mean of Group 2."
    elif alternative == 'greater':
        interpretation += "\nThere is sufficient evidence to suggest that the mean of Group 1 is significantly greater than the mean of Group 2."
  else:
    interpretation += f"Since the p-value ({p_value:.4f}) is greater than or equal to alpha ({alpha}), we fail to reject the null hypothesis."
    interpretation += "\nThere is not enough evidence to suggest a significant difference between the means of the two groups."

  return t_statistic, p_value, interpretation

# Simulate data for two groups
np.random.seed(42)  # for reproducibility
group1_scores = np.random.normal(loc=80, scale=15, size=40)  # Mean 80, Std Dev 15, size 40
group2_scores = np.random.normal(loc=85, scale=15, size=35)  # Mean 85, Std Dev 15, size 35

# Perform independent samples t-test and interpret
t_stat, p_val, interpretation = perform_independent_ttest(group1_scores, group2_scores)

print("Independent T-test Results:")
print(interpretation)

# Example with unequal variances
print("\nIndependent T-test Results (assuming unequal variances):")
t_stat_unequal, p_val_unequal, interpretation_unequal = perform_independent_ttest(group1_scores, group2_scores, equal_var=False)
print(interpretation_unequal)

# Example of a one-sided test (is the mean of group 1 less than the mean of group 2?)
print("\nOne-sided Independent T-test Results (Group 1 < Group 2):")
t_stat_less, p_val_less, interpretation_less = perform_independent_ttest(group1_scores, group2_scores, alternative='less')
interpretation_less
```

```python
# prompt: Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

def calculate_confidence_interval(sample_mean, sample_std_dev, sample_size, confidence_level=0.95, population_std_dev=None):
  """
  Calculates the confidence interval for a sample mean.

  Args:
    sample_mean (float): The mean of the sample.
    sample_std_dev (float): The standard deviation of the sample.
    sample_size (int): The number of observations in the sample.
    confidence_level (float, optional): The desired confidence level (e.g., 0.95 for 95%). Defaults to 0.95.
    population_std_dev (float, optional): The known standard deviation of the population.
                                         If None, the sample standard deviation is used and a t-distribution
                                         is assumed. Defaults to None.

  Returns:
    tuple: A tuple containing the lower bound, upper bound, and the distribution used (Z or T).
  """
  alpha = 1 - confidence_level

  if population_std_dev is not None:
    # Use Z-distribution if population standard deviation is known
    distribution = 'Z'
    # Calculate the critical Z-value for the given confidence level
    critical_value = norm.ppf(1 - alpha / 2)
    standard_error = population_std_dev / np.sqrt(sample_size)
  else:
    # Use T-distribution if population standard deviation is unknown
    distribution = 'T'
    # Degrees of freedom for the t-distribution
    degrees_of_freedom = sample_size - 1
    # Calculate the critical T-value for the given confidence level and degrees of freedom
    critical_value = stats.t.ppf(1 - alpha / 2, degrees_of_freedom)
    standard_error = sample_std_dev / np.sqrt(sample_size)

  # Calculate the margin of error
  margin_of_error = critical_value * standard_error

  # Calculate the confidence interval
  lower_bound = sample_mean - margin_of_error
  upper_bound = sample_mean + margin_of_error

  return lower_bound, upper_bound, distribution

# Example Usage:
# Assume a sample of 50 students has an average height of 170 cm and a sample
# standard deviation of 8 cm. We want to calculate a 95% confidence interval
# for the true population mean height.

sample_mean_ci = 170
sample_std_dev_ci = 8
sample_size_ci = 50
confidence_level_ci = 0.95

# Calculate CI using T-distribution (population std dev unknown)
lower_t, upper_t, distribution_t = calculate_confidence_interval(
    sample_mean_ci,
    sample_std_dev_ci,
    sample_size_ci,
    confidence_level=confidence_level_ci
)

print(f"Confidence Interval (using {distribution_t}-distribution):")
print(f"{confidence_level_ci*100}% CI: [{lower_t:.4f}, {upper_t:.4f}]")

# If the population standard deviation were known to be, say, 10 cm:
population_std_dev_known = 10

lower_z, upper_z, distribution_z = calculate_confidence_interval(
    sample_mean_ci,
    sample_std_dev_ci, # This is not strictly needed when population_std_dev is known, but the function signature requires it.
    sample_size_ci,
    confidence_level=confidence_level_ci,
    population_std_dev=population_std_dev_known
)

print(f"\nConfidence Interval (using {distribution_z}-distribution, population std dev known):")
print(f"{confidence_level_ci*100}% CI: [{lower_z:.4f}, {upper_z:.4f}]")


print("\nSignificance of the Confidence Interval:")
print("A confidence interval provides a range of plausible values for the true population mean based on sample data.")
print(f"For example, a {confidence_level_ci*100}% confidence interval means that if we were to take many random samples")
print("and calculate a confidence interval for each, we would expect approximately 95% of these intervals")
print("to contain the true population mean.")
print("\nKey points:")
print("- It provides a range of uncertainty around the point estimate (sample mean).")
print("- The width of the interval depends on the confidence level, sample size, and variability.")
print("- A wider interval indicates more uncertainty, while a narrower interval indicates more precision.")
print("- If the confidence interval does not include a specific value (e.g., a hypothesized population mean),")
print("  it provides evidence against that value at the given confidence level.")
print("  For example, if a 95% CI for mean height doesn't include 160 cm, we can be 95% confident that the true mean is not 160 cm.")
```

```python
# prompt: Write a Python program to calculate the margin of error for a given confidence level using sample data

def calculate_margin_of_error(sample_std_dev, sample_size, confidence_level=0.95, population_std_dev=None):
  """
  Calculates the margin of error for a given confidence level using sample data.

  Args:
    sample_std_dev (float): The standard deviation of the sample.
    sample_size (int): The number of observations in the sample.
    confidence_level (float, optional): The desired confidence level (e.g., 0.95 for 95%). Defaults to 0.95.
    population_std_dev (float, optional): The known standard deviation of the population.
                                         If None, the sample standard deviation is used and a t-distribution
                                         is assumed. Defaults to None.

  Returns:
    tuple: A tuple containing the calculated margin of error and the distribution used (Z or T).
  """
  alpha = 1 - confidence_level

  if population_std_dev is not None:
    # Use Z-distribution if population standard deviation is known
    distribution = 'Z'
    # Calculate the critical Z-value for the given confidence level
    critical_value = norm.ppf(1 - alpha / 2)
    standard_error = population_std_dev / np.sqrt(sample_size)
  else:
    # Use T-distribution if population standard deviation is unknown
    distribution = 'T'
    # Degrees of freedom for the t-distribution
    degrees_of_freedom = sample_size - 1
    # Calculate the critical T-value for the given confidence level and degrees of freedom
    critical_value = stats.t.ppf(1 - alpha / 2, degrees_of_freedom)
    standard_error = sample_std_dev / np.sqrt(sample_size)

  # Calculate the margin of error
  margin_of_error = critical_value * standard_error

  return margin_of_error, distribution

# Example Usage:
# Assume a sample of 50 students has a sample standard deviation of 8 cm.
# We want to calculate the margin of error for a 95% confidence level.

sample_std_dev_moe = 8
sample_size_moe = 50
confidence_level_moe = 0.95

# Calculate MOE using T-distribution (population std dev unknown)
moe_t, distribution_moe_t = calculate_margin_of_error(
    sample_std_dev_moe,
    sample_size_moe,
    confidence_level=confidence_level_moe
)

print(f"Margin of Error (using {distribution_moe_t}-distribution):")
print(f"{confidence_level_moe*100}% Margin of Error: {moe_t:.4f}")

# If the population standard deviation were known to be, say, 10 cm:
population_std_dev_known_moe = 10

moe_z, distribution_moe_z = calculate_margin_of_error(
    sample_std_dev_moe, # Sample std dev is still used for the T-case check within the function, but not in Z-case calculation.
    sample_size_moe,
    confidence_level=confidence_level_moe,
    population_std_dev=population_std_dev_known_moe
)

print(f"\nMargin of Error (using {distribution_moe_z}-distribution, population std dev known):")
print(f"{confidence_level_moe*100}% Margin of Error: {moe_z:.4f}")

print("\nInterpretation of Margin of Error:")
print("The margin of error indicates the precision of our estimate of the population mean.")
print(f"For the {confidence_level_moe*100}% confidence level, we can say that the true population mean")
print(f"is likely within the sample mean plus or minus the calculated margin of error.")
print(f"For example, if the sample mean is 170 cm and the MOE is {moe_t:.4f} cm, the 95% CI is [170 - {moe_t:.4f}, 170 + {moe_t:.4f}].")
print("\nFactors affecting Margin of Error:")
print("- Confidence Level: Higher confidence levels result in larger margins of error.")
print("- Sample Size: Larger sample sizes result in smaller margins of error.")
print("- Variability (Standard Deviation): Higher variability in the data results in larger margins of error.")
```

```python
# prompt:  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

# Implement a Bayesian inference method using Bayes' Theorem

def bayesian_inference(prior_prob_a, prob_b_given_a, prob_b_given_not_a):
  """
  Performs a simple Bayesian inference calculation using Bayes' Theorem.

  Explains the process of updating the probability of event A given evidence B.

  Args:
    prior_prob_a (float): The prior probability of event A, P(A).
    prob_b_given_a (float): The probability of event B given A, P(B|A) (Likelihood).
    prob_b_given_not_a (float): The probability of event B given not A, P(B|~A).

  Returns:
    float: The posterior probability of event A given B, P(A|B).
  """
  # Ensure probabilities are valid (between 0 and 1)
  if not (0 <= prior_prob_a <= 1 and 0 <= prob_b_given_a <= 1 and 0 <= prob_b_given_not_a <= 1):
    print("Error: Probabilities must be between 0 and 1.")
    return None

  # Probability of not A
  prior_prob_not_a = 1 - prior_prob_a

  # Probability of B (using the law of total probability): P(B) = P(B|A)P(A) + P(B|~A)P(~A)
  prob_b = (prob_b_given_a * prior_prob_a) + (prob_b_given_not_a * prior_prob_not_a)

  # Handle the case where P(B) is zero to avoid division by zero
  if prob_b == 0:
      print("Error: Probability of evidence B is zero. Cannot perform inference.")
      return None

  # Apply Bayes' Theorem: P(A|B) = [P(B|A) * P(A)] / P(B)
  posterior_prob_a_given_b = (prob_b_given_a * prior_prob_a) / prob_b

  print("\nBayesian Inference Process:")
  print(f"1. Start with a Prior Probability of A, P(A): {prior_prob_a:.4f}")
  print("   This is your initial belief about the probability of event A before observing any evidence.")
  print(f"2. Observe new Evidence B.")
  print(f"3. Calculate the Likelihood of Evidence B given A, P(B|A): {prob_b_given_a:.4f}")
  print("   This is the probability of observing the evidence if event A is true.")
  print(f"4. Calculate the Probability of Evidence B given Not A, P(B|~A): {prob_b_given_not_a:.4f}")
  print("   This is the probability of observing the evidence if event A is false.")
  print(f"5. Calculate the Marginal Probability of Evidence B, P(B): {prob_b:.4f}")
  print("   This is the overall probability of observing the evidence, considering both cases where A is true or false.")
  print("   P(B) = P(B|A)P(A) + P(B|~A)P(~A)")
  print(f"6. Apply Bayes' Theorem to calculate the Posterior Probability of A given B, P(A|B): {posterior_prob_a_given_b:.4f}")
  print("   P(A|B) = [P(B|A) * P(A)] / P(B)")
  print(f"   This is your updated belief about the probability of event A after observing the evidence B.")

  return posterior_prob_a_given_b

# Example Usage: Medical Diagnosis
# Let A be the event that a patient has a rare disease.
# Let B be the event that the patient tests positive for the disease.

# Prior probability of having the disease (prevalence)
prior_prob_disease = 0.01 # 1% of the population has the disease, P(A)

# Probability of testing positive given the patient has the disease (sensitivity)
prob_positive_given_disease = 0.95 # The test is 95% sensitive, P(B|A)

# Probability of testing positive given the patient does NOT have the disease (false positive rate)
prob_positive_given_no_disease = 0.02 # The test has a 2% false positive rate, P(B|~A)

# Now, a patient tests positive. What is the probability they actually have the disease?
posterior_prob_disease_given_positive = bayesian_inference(
    prior_prob_disease,
    prob_positive_given_disease,
    prob_positive_given_no_disease
)

if posterior_prob_disease_given_positive is not None:
  print(f"\nResult: The posterior probability of having the disease given a positive test result is: {posterior_prob_disease_given_positive:.4f}")
  print("Interpretation: Although the test is 95% sensitive, the relatively low prevalence of the disease and the 2% false positive rate mean that a positive test result only increases the probability of having the disease from 1% (prior) to approximately {:.2f}% (posterior).".format(posterior_prob_disease_given_positive * 100))
  print("This highlights how prior beliefs and the reliability of the evidence (likelihoods) impact the updated probability.")
```

```python
# prompt: Perform a Chi-square test for independence between two categorical variables in Python

from scipy.stats import chi2_contingency
import pandas as pd

def perform_chi_square_test(data, alpha=0.05):
  """
  Performs a Chi-square test for independence on a contingency table.

  Args:
    data (list of lists or np.array or pd.DataFrame): A contingency table
      representing the observed frequencies of two categorical variables.
    alpha (float, optional): The significance level (alpha). Defaults to 0.05.

  Returns:
    tuple: A tuple containing the chi-square statistic, p-value, degrees of
           freedom, and interpretation.
  """
  # Perform the chi-square test
  chi2_statistic, p_value, dof, expected_frequencies = chi2_contingency(data)

  # Interpret the results
  interpretation = f"Chi-square Statistic: {chi2_statistic:.4f}\n"
  interpretation += f"P-value: {p_value:.4f}\n"
  interpretation += f"Degrees of Freedom: {dof}\n"
  interpretation += "Expected Frequencies:\n"
  # Convert expected frequencies to a DataFrame for better display
  interpretation += pd.DataFrame(expected_frequencies).to_string(index=False, float_format='{:.2f}'.format)
  interpretation += "\n\n"

  if p_value < alpha:
    interpretation += f"Since the p-value ({p_value:.4f}) is less than alpha ({alpha}), we reject the null hypothesis."
    interpretation += "\nThere is sufficient evidence to suggest a significant association (dependence) between the two categorical variables."
  else:
    interpretation += f"Since the p-value ({p_value:.4f}) is greater than or equal to alpha ({alpha}), we fail to reject the null hypothesis."
    interpretation += "\nThere is not enough evidence to suggest a significant association (independence assumed) between the two categorical variables."

  return chi2_statistic, p_value, dof, interpretation

# Example Usage:
# Let's consider an example where we want to test if there is a relationship
# between gender and preference for a certain type of movie genre (e.g., Sci-Fi vs Romance).

# Create a contingency table of observed frequencies:
# Rows represent Gender (Male, Female)
# Columns represent Movie Genre (Sci-Fi, Romance)
observed_data = np.array([
    [60, 30],  # Males who prefer Sci-Fi, Males who prefer Romance
    [40, 50]   # Females who prefer Sci-Fi, Females who prefer Romance
])

print("Observed Frequencies:")
print(observed_data)

# Perform the Chi-square test for independence
chi2_stat, p_val, degrees_of_freedom, interpretation = perform_chi_square_test(observed_data)

print("\n--- Chi-square Test for Independence Results ---")
interpretation
```

```python
# prompt:  Implement an F-test using Python to compare the variances of two random samples

def perform_f_test(sample1_data, sample2_data, alpha=0.05, alternative='two-sided'):
  """
  Performs an F-test to compare the variances of two random samples.

  Args:
    sample1_data (list or np.array): Data for the first sample.
    sample2_data (list or np.array): Data for the second sample.
    alpha (float, optional): The significance level (alpha). Defaults to 0.05.
    alternative (str, optional): The alternative hypothesis. Can be 'two-sided',
                                 'less' (variance of sample 1 < variance of sample 2),
                                 or 'greater' (variance of sample 1 > variance of sample 2).
                                 Defaults to 'two-sided'.

  Returns:
    tuple: A tuple containing the F-statistic, p-value, and interpretation.
  """
  # Calculate sample variances
  sample1_variance = np.var(sample1_data, ddof=1) # ddof=1 for sample variance
  sample2_variance = np.var(sample2_data, ddof=1)

  # Ensure the larger variance is in the numerator for one-sided 'greater' test,
  # or for convenience in two-sided as well (though scipy handles this).
  # For one-sided 'less', we need sample1_variance in the numerator.
  if alternative == 'greater' and sample1_variance < sample2_variance:
      # Swap samples to ensure larger variance is in numerator for 'greater' test logic
      sample1_variance, sample2_variance = sample2_variance, sample1_variance
      sample1_size, sample2_size = len(sample2_data), len(sample1_data) # Swap sizes too
      is_swapped = True
      print("Note: Samples swapped for 'greater' alternative to ensure larger variance is in numerator for F-statistic calculation.")
  elif alternative == 'less' and sample1_variance > sample2_variance:
      # This case isn't strictly necessary as the F-statistic will be < 1,
      # but keeping for clarity on the calculation.
      sample1_size, sample2_size = len(sample1_data), len(sample2_data)
      is_swapped = False
  else:
      sample1_size, sample2_size = len(sample1_data), len(sample2_data)
      is_swapped = False

  # Degrees of freedom
  df1 = sample1_size - 1
  df2 = sample2_size - 1

  if df1 <= 0 or df2 <= 0:
      return None, None, "Error: Sample size must be at least 2 for each sample to calculate variance."

  # Calculate the F-statistic
  if sample2_variance == 0:
      return None, None, "Error: Variance of sample 2 is zero."
  f_statistic = sample1_variance / sample2_variance

  # Calculate the p-value
  if alternative == 'two-sided':
    # For a two-sided test, we find the probability in the tail and multiply by 2.
    # We use the cumulative distribution function (cdf) to find the probability
    # of observing an F-statistic as extreme or more extreme than the calculated one.
    # Since the F-distribution is not symmetric, we need to consider both tails
    # based on the actual F-statistic value relative to 1.
    if f_statistic >= 1:
        p_value = 2 * (1 - stats.f.cdf(f_statistic, df1, df2))
    else: # f_statistic < 1
        p_value = 2 * stats.f.cdf(f_statistic, df1, df2)

  elif alternative == 'greater':
    # For a one-sided 'greater' test, we find the probability in the right tail.
    p_value = 1 - stats.f.cdf(f_statistic, df1, df2)

  elif alternative == 'less':
    # For a one-sided 'less' test, we find the probability in the left tail.
    # If we didn't swap samples, f_statistic might be > 1 even for 'less'.
    # The p-value is the probability of getting an F-statistic <= the observed one.
    p_value = stats.f.cdf(f_statistic, df1, df2)

  else:
    return None, None, "Invalid alternative hypothesis. Choose 'two-sided', 'less', or 'greater'."


  # Interpret the results
  interpretation = f"F-statistic: {f_statistic:.4f}\n"
  interpretation += f"P-value: {p_value:.4f}\n"
  interpretation += f"Degrees of Freedom (df1, df2): ({df1}, {df2})\n"
  if is_swapped:
      interpretation += "(Note: Sample data order was conceptually swapped for F-statistic calculation to ensure larger variance in numerator for 'greater' test interpretation.)\n"


  if p_value < alpha:
    interpretation += f"Since the p-value ({p_value:.4f}) is less than alpha ({alpha}), we reject the null hypothesis."
    if alternative == 'two-sided':
        interpretation += "\nThere is sufficient evidence to suggest that the variances of the two populations are significantly different."
    elif alternative == 'greater':
         interpretation += "\nThere is sufficient evidence to suggest that the variance of Sample 1 is significantly greater than the variance of Sample 2."
    elif alternative == 'less':
        interpretation += "\nThere is sufficient evidence to suggest that the variance of Sample 1 is significantly less than the variance of Sample 2."

  else:
    interpretation += f"Since the p-value ({p_value:.4f}) is greater than or equal to alpha ({alpha}), we fail to reject the null hypothesis."
    interpretation += "\nThere is not enough evidence to suggest that the variances of the two populations are significantly different."

  return f_statistic, p_value, interpretation

# Example Usage:
# Simulate two samples with potentially different variances

np.random.seed(42) # for reproducibility
sample_a = np.random.normal(loc=50, scale=10, size=30) # Mean 50, Std Dev 10
sample_b = np.random.normal(loc=55, scale=15, size=35) # Mean 55, Std Dev 15 (larger variance)

print("--- F-test to compare variances ---")
print(f"Sample A Variance: {np.var(sample_a, ddof=1):.4f}")
print(f"Sample B Variance: {np.var(sample_b, ddof=1):.4f}")

# Perform a two-sided F-test
f_stat, p_val, interpretation = perform_f_test(sample_a, sample_b, alpha=0.05)
print("\nTwo-sided F-test results:")
print(interpretation)

# Perform a one-sided F-test (Is variance of B greater than variance of A?)
f_stat_greater, p_val_greater, interpretation_greater = perform_f_test(sample_b, sample_a, alpha=0.05, alternative='greater')
print("\nOne-sided F-test results (Is variance of B > variance of A?):")
print(interpretation_greater)

# Perform a one-sided F-test (Is variance of A less than variance of B?)
f_stat_less, p_val_less, interpretation_less = perform_f_test(sample_a, sample_b, alpha=0.05, alternative='less')
print("\nOne-sided F-test results (Is variance of A < variance of B?):")
interpretation_less
```

```python
# prompt:  Write a Python program to perform an ANOVA test to compare means between multiple groups and
# interpret the results

from scipy import stats
import pandas as pd
import numpy as np

def perform_one_way_anova(group_data, alpha=0.05):
  """
  Performs a one-way ANOVA test and interprets the results.

  Args:
    group_data (list of arrays or list of lists): A list where each element
                is an array or list of data for a single group.
    alpha (float, optional): The significance level (alpha). Defaults to 0.05.

  Returns:
    tuple: A tuple containing the F-statistic, p-value, and interpretation.
  """
  # Check if there are at least two groups
  if len(group_data) < 2:
    return None, None, "ANOVA requires at least two groups."

  # Perform one-way ANOVA
  # The f_oneway function requires the data for each group as separate arguments.
  f_statistic, p_value = stats.f_oneway(*group_data)

  # Interpret the results
  interpretation = f"F-statistic: {f_statistic:.4f}\n"
  interpretation += f"P-value: {p_value:.4f}\n"

  if p_value < alpha:
    interpretation += f"Since the p-value ({p_value:.4f}) is less than alpha ({alpha}), we reject the null hypothesis."
    interpretation += "\nConclusion: There is sufficient evidence to suggest that at least one group mean is significantly different from the others."
    interpretation += "\n(Further post-hoc tests are needed to determine which specific group means differ)."
  else:
    interpretation += f"Since the p-value ({p_value:.4f}) is greater than or equal to alpha ({alpha}), we fail to reject the null hypothesis."
    interpretation += "\nConclusion: There is not enough evidence to suggest a significant difference between the means of the groups."

  return f_statistic, p_value, interpretation

# Example Usage:
# Assume we have data on the test scores of students from three different
# teaching methods (Group A, Group B, Group C). We want to know if the
# average test scores are significantly different across the three methods.

# Simulate data for three groups
np.random.seed(42) # for reproducibility
group_a_scores = np.random.normal(loc=75, scale=10, size=30)
group_b_scores = np.random.normal(loc=80, scale=10, size=30)
group_c_scores = np.random.normal(loc=70, scale=10, size=30)

# Store the data in a list of arrays
all_group_data = [group_a_scores, group_b_scores, group_c_scores]

# Perform One-Way ANOVA and interpret
f_stat, p_val, interpretation = perform_one_way_anova(all_group_data, alpha=0.05)

print("--- One-Way ANOVA Results ---")
interpretation

# To see which specific groups differ after a significant ANOVA result,
# you would typically perform post-hoc tests (e.g., Tukey's HSD).
# Here's an example using the `statsmodels` library (requires installation: !pip install statsmodels)
# from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Combine all data and create a group label
# all_scores = np.concatenate(all_group_data)
# group_labels = ['A'] * len(group_a_scores) + ['B'] * len(group_b_scores) + ['C'] * len(group_c_scores)

# Perform Tukey's HSD post-hoc test
# tukey_results = pairwise_tukeyhsd(endog=all_scores, groups=group_labels, alpha=0.05)
# print("\nTukey's HSD Post-hoc Test Results:")
# print(tukey_results)
```

```python
# prompt: Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

from scipy.stats import shapiro, levene

def check_anova_assumptions(groups_data, alpha=0.05):
  """
  Checks the assumptions for one-way ANOVA:
  1. Normality: Data in each group is normally distributed.
  2. Independence: Observations within and between groups are independent.
     (This assumption is typically met by experimental design, not statistically tested here).
  3. Equal Variances (Homoscedasticity): Variances are approximately equal across groups.

  Args:
    groups_data (list of arrays or list of lists): A list where each element
                 is an array or list of data for a single group.
    alpha (float, optional): The significance level for the normality and
             equal variance tests. Defaults to 0.05.

  Returns:
    dict: A dictionary containing the results of the assumption checks.
  """
  results = {}

  # 1. Check for Normality (using Shapiro-Wilk test for each group)
  print("\n--- Normality Check (Shapiro-Wilk Test) ---")
  results['normality'] = {}
  for i, group_data in enumerate(groups_data):
    group_name = f'Group {i+1}'
    if len(group_data) < 3:
        print(f"Warning: Sample size for {group_name} is too small (<3) to perform Shapiro-Wilk test.")
        results['normality'][group_name] = "Sample size too small"
        continue
    if len(group_data) > 5000:
         print(f"Warning: Sample size for {group_name} is large (>5000), Shapiro-Wilk test may be less accurate. Consider other tests like D'Agostino-Pearson.")
         # Still perform Shapiro-Wilk for demonstration
         pass


    stat, p_value_norm = shapiro(group_data)
    results['normality'][group_name] = {'statistic': stat, 'p_value': p_value_norm}
    print(f"{group_name}: Shapiro-Wilk Statistic={stat:.4f}, P-value={p_value_norm:.4f}")
    if p_value_norm < alpha:
      print(f"  Assumption VIOLATED: Data in {group_name} is likely not normally distributed (p < {alpha}).")
    else:
      print(f"  Assumption MET: Data in {group_name} appears normally distributed (p >= {alpha}).")

  # 2. Check for Independence
  print("\n--- Independence Check ---")
  results['independence'] = "Requires experimental design verification. Not statistically tested here."
  print("This assumption (independence of observations within and between groups) is typically met by the study design (e.g., random assignment, no repeated measures on the same individuals in different groups). You should confirm this based on how the data was collected.")


  # 3. Check for Equal Variances (using Levene's test)
  print("\n--- Equal Variances Check (Levene's Test) ---")
  if len(groups_data) < 2:
      print("Error: Levene's test requires at least two groups.")
      results['equal_variances'] = "Requires at least two groups"
  else:
      # Levene's test takes the data for each group as separate arguments
      stat_levene, p_value_levene = levene(*groups_data)
      results['equal_variances'] = {'statistic': stat_levene, 'p_value': p_value_levene}
      print(f"Levene's Test Statistic={stat_levene:.4f}, P-value={p_value_levene:.4f}")
      if p_value_levene < alpha:
        print(f"  Assumption VIOLATED: Variances across groups are significantly different (p < {alpha}).")
        print("  Consider using Welch's ANOVA (if available in your ANOVA function) or transforming the data.")
      else:
        print(f"  Assumption MET: Variances across groups appear approximately equal (p >= {alpha}).")

  return results

# Example Usage:
# Use the simulated data from the previous ANOVA example
# Simulate data for three groups again for clarity in this function's example
np.random.seed(43) # Use a different seed for variety
group1_scores_assump = np.random.normal(loc=75, scale=10, size=30)
group2_scores_assump = np.random.normal(loc=80, scale=12, size=35) # Slightly different mean and variance
group3_scores_assump = np.random.normal(loc=70, scale=8, size=25)   # Different mean and variance

all_group_data_assump = [group1_scores_assump, group2_scores_assump, group3_scores_assump]

assumption_check_results = check_anova_assumptions(all_group_data_assump, alpha=0.05)

print("\n--- Summary of ANOVA Assumption Check Results ---")
# You can print the results dictionary or process it further
# print(assumption_check_results) # Uncomment to print the raw dictionary

# More structured summary printing
print("Normality Check per Group:")
for group, result in assumption_check_results['normality'].items():
    if isinstance(result, str):
        print(f"  {group}: {result}")
    else:
        print(f"  {group}: P-value = {result['p_value']:.4f} (Assumption {'MET' if result['p_value'] >= 0.05 else 'VIOLATED'})")

print("\nIndependence Check:")
print(f"  Status: {assumption_check_results['independence']}")

print("\nEqual Variances (Homoscedasticity) Check:")
levene_result = assumption_check_results['equal_variances']
if isinstance(levene_result, str):
     print(f"  Status: {levene_result}")
else:
    print(f"  Levene's Test P-value = {levene_result['p_value']:.4f} (Assumption {'MET' if levene_result['p_value'] >= 0.05 else 'VIOLATED'})")

```

```python
# prompt:  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

def perform_chi_square_variance_test(sample_data, hypothesized_variance, alpha=0.05, alternative='two-sided'):
  """
  Performs a Chi-square test for population variance and interprets the results.

  Args:
    sample_data (list or np.array): The data from the sample.
    hypothesized_variance (float): The hypothesized value for the population variance (œÉ¬≤‚ÇÄ).
    alpha (float, optional): The significance level (alpha). Defaults to 0.05.
    alternative (str, optional): The alternative hypothesis. Can be 'two-sided' (œÉ¬≤ ‚â† œÉ¬≤‚ÇÄ),
                                 'less' (œÉ¬≤ < œÉ¬≤‚ÇÄ), or 'greater' (œÉ¬≤ > œÉ¬≤‚ÇÄ).
                                 Defaults to 'two-sided'.

  Returns:
    tuple: A tuple containing the Chi-square statistic, p-value, and interpretation.
  """
  sample_data = np.array(sample_data)
  sample_size = len(sample_data)

  if sample_size < 2:
    return None, None, "Error: Sample size must be at least 2 to calculate variance."

  if hypothesized_variance <= 0:
      return None, None, "Error: Hypothesized variance must be positive."

  # Calculate the sample variance
  sample_variance = np.var(sample_data, ddof=1) # ddof=1 for sample variance (s¬≤)

  # Calculate the Chi-square test statistic
  # Formula: œá¬≤ = (n - 1) * s¬≤ / œÉ¬≤‚ÇÄ
  chi2_statistic = (sample_size - 1) * sample_variance / hypothesized_variance

  # Degrees of freedom
  degrees_of_freedom = sample_size - 1

  # Calculate the p-value
  if alternative == 'two-sided':
    # For a two-sided test, we need the area in both tails.
    # We find the probability of observing a chi-square statistic as extreme
    # or more extreme than the calculated one.
    # If the statistic is in the right tail, the p-value is 2 * (1 - cdf(chi2_stat)).
    # If the statistic is in the left tail, the p-value is 2 * cdf(chi2_stat).
    # A simpler way is to find the minimum of the tail probabilities multiplied by 2.
    p_value = min(chi2.cdf(chi2_statistic, degrees_of_freedom), 1 - chi2.cdf(chi2_statistic, degrees_of_freedom)) * 2
  elif alternative == 'greater':
    # For a one-sided 'greater' test, we find the area in the right tail.
    p_value = 1 - chi2.cdf(chi2_statistic, degrees_of_freedom)
  elif alternative == 'less':
    # For a one-sided 'less' test, we find the area in the left tail.
    p_value = chi2.cdf(chi2_statistic, degrees_of_freedom)
  else:
    return None, None, "Invalid alternative hypothesis. Choose 'two-sided', 'less', or 'greater'."

  # Interpret the results
  interpretation = f"Chi-square Statistic: {chi2_statistic:.4f}\n"
  interpretation += f"P-value: {p_value:.4f}\n"
  interpretation += f"Degrees of Freedom: {degrees_of_freedom}\n"
  interpretation += f"Sample Variance (s¬≤): {sample_variance:.4f}\n"
  interpretation += f"Hypothesized Population Variance (œÉ¬≤‚ÇÄ): {hypothesized_variance:.4f}\n"
  interpretation += f"Alternative Hypothesis: Population Variance {alternative} {hypothesized_variance:.4f}\n"


  if p_value < alpha:
    interpretation += f"Since the p-value ({p_value:.4f}) is less than alpha ({alpha}), we reject the null hypothesis."
    if alternative == 'two-sided':
        interpretation += f"\nConclusion: There is sufficient evidence to suggest that the population variance is significantly different from the hypothesized value of {hypothesized_variance:.4f}."
    elif alternative == 'greater':
         interpretation += f"\nConclusion: There is sufficient evidence to suggest that the population variance is significantly greater than the hypothesized value of {hypothesized_variance:.4f}."
    elif alternative == 'less':
        interpretation += f"\nConclusion: There is sufficient evidence to suggest that the population variance is significantly less than the hypothesized value of {hypothesized_variance:.4f}."

  else:
    interpretation += f"Since the p-value ({p_value:.4f}) is greater than or equal to alpha ({alpha}), we fail to reject the null hypothesis."
    interpretation += f"\nConclusion: There is not enough evidence to suggest that the population variance is significantly different from the hypothesized value of {hypothesized_variance:.4f}."

  return chi2_statistic, p_value, interpretation

# Example Usage:
# Assume we have a sample of 20 measurements and we want to test if the population
# variance is significantly different from a hypothesized value of 15.

# Simulate sample data
np.random.seed(42) # for reproducibility
# Simulate data from a population with actual variance around 20
sample_data_var_test = np.random.normal(loc=100, scale=np.sqrt(20), size=20)

hypothesized_population_variance = 15
alpha_var_test = 0.05

print("--- Chi-square Test for Population Variance ---")
print(f"Sample Data (first 5 values): {sample_data_var_test[:5]}")
print(f"Sample Size: {len(sample_data_var_test)}")
print(f"Hypothesized Population Variance (œÉ¬≤‚ÇÄ): {hypothesized_population_variance}")
print(f"Significance Level (alpha): {alpha_var_test}")
print(f"Sample Variance (s¬≤): {np.var(sample_data_var_test, ddof=1):.4f}")


# Perform a two-sided test
chi2_stat, p_val, interpretation = perform_chi_square_variance_test(
    sample_data_var_test,
    hypothesized_population_variance,
    alpha=alpha_var_test,
    alternative='two-sided'
)
print("\nTwo-sided test results (H‚ÇÄ: œÉ¬≤ = 15, H‚ÇÅ: œÉ¬≤ ‚â† 15):")
print(interpretation)


# Perform a one-sided 'greater' test
chi2_stat_greater, p_val_greater, interpretation_greater = perform_chi_square_variance_test(
    sample_data_var_test,
    hypothesized_population_variance,
    alpha=alpha_var_test,
    alternative='greater'
)
print("\nOne-sided test results (H‚ÇÄ: œÉ¬≤ = 15, H‚ÇÅ: œÉ¬≤ > 15):")
print(interpretation_greater)


# Perform a one-sided 'less' test (example with a hypothesized variance likely larger than actual)
hypothesized_population_variance_less = 30
print(f"\nOne-sided test results (H‚ÇÄ: œÉ¬≤ = 30, H‚ÇÅ: œÉ¬≤ < 30):")
chi2_stat_less, p_val_less, interpretation_less = perform_chi_square_variance_test(
    sample_data_var_test,
    hypothesized_population_variance_less,
    alpha=alpha_var_test,
    alternative='less'
)
print(interpretation_less)


print("\n--- Interpretation Summary ---")
print("The Chi-square test for variance is used to determine if a sample variance is significantly different from a known or hypothesized population variance (œÉ¬≤‚ÇÄ).")
print("Null Hypothesis (H‚ÇÄ): The population variance is equal to the hypothesized value (œÉ¬≤ = œÉ¬≤‚ÇÄ).")
print("Alternative Hypothesis (H‚ÇÅ): The population variance is different from (‚â†), greater than (>), or less than (<) the hypothesized value.")
print("The test statistic follows a Chi-square distribution with n-1 degrees of freedom, where n is the sample size.")
print("A small p-value (typically < alpha) leads to rejecting the null hypothesis, indicating that the population variance is likely different from the hypothesized value.")
print("A large p-value (‚â• alpha) means we fail to reject the null hypothesis, suggesting there isn't enough evidence to conclude the population variance is different from the hypothesized value.")

# Visualize the Chi-square distribution and the test statistic location
degrees_of_freedom_vis = len(sample_data_var_test) - 1
chi2_statistic_vis = (len(sample_data_var_test) - 1) * np.var(sample_data_var_test, ddof=1) / hypothesized_population_variance # Use the value from the first test

# Generate points on the x-axis (Chi-square values)
# Use a range that makes sense for the degrees of freedom
x_vis = np.linspace(0, chi2.ppf(0.999, degrees_of_freedom_vis), 200)

# Calculate the probability density function (PDF)
pdf_vis = chi2.pdf(x_vis, degrees_of_freedom_vis)

plt.figure(figsize=(10, 6))
plt.plot(x_vis, pdf_vis, label=f'Chi-square Distribution (df = {degrees_of_freedom_vis})')

# Plot the calculated Chi-square statistic
plt.axvline(chi2_statistic_vis, color='red', linestyle='dashed', linewidth=2, label=f'Calculated œá¬≤: {chi2_statistic_vis:.4f}')

# Add titles and labels
plt.title('Chi-square Test for Population Variance')
plt.xlabel('Chi-square Value')
plt.ylabel('Probability Density')

# Optionally shade the rejection region for the two-sided test
alpha_vis = alpha_var_test
critical_left = chi2.ppf(alpha_vis / 2, degrees_of_freedom_vis)
critical_right = chi2.ppf(1 - alpha_vis / 2, degrees_of_freedom_vis)

x_fill_left_vis = x_vis[x_vis <= critical_left]
pdf_fill_left_vis = pdf_vis[x_vis <= critical_left]
plt.fill_between(x_fill_left_vis, pdf_fill_left_vis, color='orange', alpha=0.5, label=f'Rejection Region (Left, Œ±/2)')

x_fill_right_vis = x_vis[x_vis >= critical_right]
pdf_fill_right_vis = pdf_vis[x_vis >= critical_right]
plt.fill_between(x_fill_right_vis, pdf_fill_right_vis, color='orange', alpha=0.5, label=f'Rejection Region (Right, Œ±/2)')


plt.legend()
plt.grid(True)
plt.show()

print("\nVisual Interpretation:")
print(f"- The plot shows the Chi-square distribution with {degrees_of_freedom_vis} degrees of freedom.")
print(f"- The dashed red line indicates the location of the calculated Chi-square test statistic ({chi2_statistic_vis:.4f}).")
print(f"- For a two-sided test at alpha={alpha_vis}, the orange shaded regions represent the rejection regions in both tails.")
print(f"- The critical values are approximately {critical_left:.4f} (left) and {critical_right:.4f} (right).")
if chi2_statistic_vis < critical_left or chi2_statistic_vis > critical_right:
    print(f"- Since the calculated Chi-square statistic ({chi2_statistic_vis:.4f}) falls within the rejection region, we reject the null hypothesis.")
else:
    print(f"- Since the calculated Chi-square statistic ({chi2_statistic_vis:.4f}) falls outside the rejection region, we fail to reject the null hypothesis.")
print(f"  (This visual interpretation aligns with the p-value results from the two-sided test: p-value {p_val:.4f}).")

```

```python
# prompt:  Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

def perform_z_test_proportions(sample1_successes, sample1_size, sample2_successes, sample2_size, alpha=0.05, alternative='two-sided'):
  """
  Performs a Z-test for comparing proportions between two independent groups
  and interprets the results.

  Args:
    sample1_successes (int): The number of successes in sample 1.
    sample1_size (int): The total number of observations in sample 1.
    sample2_successes (int): The number of successes in sample 2.
    sample2_size (int): The total number of observations in sample 2.
    alpha (float, optional): The significance level (alpha). Defaults to 0.05.
    alternative (str, optional): The alternative hypothesis. Can be 'two-sided',
                                 'less' (proportion 1 < proportion 2), or
                                 'greater' (proportion 1 > proportion 2).
                                 Defaults to 'two-sided'.

  Returns:
    tuple: A tuple containing the Z-statistic, p-value, and interpretation.
  """
  # Calculate sample proportions
  p1 = sample1_successes / sample1_size
  p2 = sample2_successes / sample2_size

  # Calculate the pooled proportion (p-hat)
  pooled_proportion = (sample1_successes + sample2_successes) / (sample1_size + sample2_size)

  # Calculate the standard error of the difference in proportions
  # Formula: SE = sqrt(p_hat * (1 - p_hat) * (1/n1 + 1/n2))
  standard_error = np.sqrt(pooled_proportion * (1 - pooled_proportion) * (1/sample1_size + 1/sample2_size))

  # Calculate the Z-statistic
  # Formula: Z = (p1 - p2) / SE
  z_statistic = (p1 - p2) / standard_error

  # Calculate the p-value
  if alternative == 'two-sided':
    # For a two-sided test, the p-value is the area in both tails.
    # We multiply the tail probability of the absolute Z-statistic by 2.
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
  elif alternative == 'less':
    # For a one-sided 'less' test, the p-value is the area in the left tail.
    p_value = stats.norm.cdf(z_statistic)
  elif alternative == 'greater':
    # For a one-sided 'greater' test, the p-value is the area in the right tail.
    p_value = 1 - stats.norm.cdf(z_statistic)
  else:
    return None, None, "Invalid alternative hypothesis. Choose 'two-sided', 'less', or 'greater'."

  # Interpret the results
  interpretation = f"Sample 1 Proportion: {p1:.4f} ({sample1_successes}/{sample1_size})\n"
  interpretation += f"Sample 2 Proportion: {p2:.4f} ({sample2_successes}/{sample2_size})\n"
  interpretation += f"Pooled Proportion: {pooled_proportion:.4f}\n"
  interpretation += f"Standard Error: {standard_error:.4f}\n"
  interpretation += f"Z-statistic: {z_statistic:.4f}\n"
  interpretation += f"P-value: {p_value:.4f}\n"
  interpretation += f"Significance Level (alpha): {alpha}\n"
  interpretation += f"Alternative Hypothesis: Proporton 1 {alternative.replace('two-sided', '‚â†').replace('less', '<').replace('greater', '>')} Proportion 2\n\n"


  if p_value < alpha:
    interpretation += f"Since the p-value ({p_value:.4f}) is less than alpha ({alpha}), we reject the null hypothesis."
    if alternative == 'two-sided':
        interpretation += "\nConclusion: There is sufficient evidence to suggest a significant difference between the population proportions of the two groups."
    elif alternative == 'less':
        interpretation += "\nConclusion: There is sufficient evidence to suggest that the population proportion of Group 1 is significantly less than that of Group 2."
    elif alternative == 'greater':
        interpretation += "\nConclusion: There is sufficient evidence to suggest that the population proportion of Group 1 is significantly greater than that of Group 2."
  else:
    interpretation += f"Since the p-value ({p_value:.4f}) is greater than or equal to alpha ({alpha}), we fail to reject the null hypothesis."
    interpretation += "\nConclusion: There is not enough evidence to suggest a significant difference between the population proportions of the two groups."

  return z_statistic, p_value, interpretation

# Example Usage:
# A marketing company wants to compare the click-through rate (CTR) of two different ad campaigns.
# Campaign A was shown to 500 people, resulting in 80 clicks.
# Campaign B was shown to 600 people, resulting in 100 clicks.
# Is there a significant difference in CTR between the two campaigns at a 0.05 significance level?

sample1_successes = 80  # Clicks for Campaign A
sample1_size = 500    # Views for Campaign A

sample2_successes = 100 # Clicks for Campaign B
sample2_size = 600    # Views for Campaign B

alpha_prop = 0.05

print("--- Z-test for Comparing Two Proportions ---")
z_stat_prop, p_val_prop, interpretation_prop = perform_z_test_proportions(
    sample1_successes, sample1_size,
    sample2_successes, sample2_size,
    alpha=alpha_prop,
    alternative='two-sided' # Testing for any difference
)
print(interpretation_prop)

# Example with a one-sided test: Is the proportion for Sample 1 greater than Sample 2?
print("\nOne-sided Z-test (Proportion 1 > Proportion 2):")
z_stat_prop_greater, p_val_prop_greater, interpretation_prop_greater = perform_z_test_proportions(
    sample1_successes, sample1_size,
    sample2_successes, sample2_size,
    alpha=alpha_prop,
    alternative='greater'
)
print(interpretation_prop_greater)

# Example with a one-sided test: Is the proportion for Sample 1 less than Sample 2?
print("\nOne-sided Z-test (Proportion 1 < Proportion 2):")
z_stat_prop_less, p_val_prop_less, interpretation_prop_less = perform_z_test_proportions(
    sample1_successes, sample1_size,
    sample2_successes, sample2_size,
    alpha=alpha_prop,
    alternative='less'
)
interpretation_prop_less

```

```python
# prompt: Perform a Chi-square test for goodness of fit with simulated data and analyze the results

# Simulate data for a Chi-square goodness-of-fit test
# We want to test if observed frequencies from a single sample fit a hypothesized distribution.
# Example: Testing if a die is fair.
# Hypothesized distribution: Each face (1-6) has an equal probability (1/6).

# Number of trials (e.g., total number of die rolls)
total_rolls = 120

# Hypothesized probabilities for each outcome (for a fair die)
hypothesized_probabilities = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
# Make sure probabilities sum to 1
if not np.isclose(np.sum(hypothesized_probabilities), 1.0):
    print("Warning: Hypothesized probabilities do not sum to 1. Adjusting.")
    hypothesized_probabilities = hypothesized_probabilities / np.sum(hypothesized_probabilities)


# Calculate the expected frequencies based on the total number of trials
# Expected Frequency = Total Trials * Hypothesized Probability
expected_frequencies_gof = total_rolls * hypothesized_probabilities

# Simulate observed frequencies (e.g., actual counts from rolling the die)
# Let's simulate a slightly unfair die to see if the test detects it
observed_frequencies = np.array([15, 20, 25, 18, 22, 20]) # Sum should be total_rolls (120)

# Ensure the sum of observed frequencies matches the total number of trials
if np.sum(observed_frequencies) != total_rolls:
    print(f"Warning: Sum of observed frequencies ({np.sum(observed_frequencies)}) does not match total trials ({total_rolls}).")
    # For the test to be valid, the sums must match. You might need to rescale
    # observed frequencies or check data collection.

# Perform the Chi-square goodness-of-fit test
# The stats.chisquare function takes observed and expected frequencies directly
chi2_statistic_gof, p_value_gof = stats.chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies_gof)

# Set significance level
alpha_gof = 0.05

# Degrees of freedom for goodness-of-fit is k - 1, where k is the number of categories
dof_gof = len(observed_frequencies) - 1

# Interpret the results
print("--- Chi-square Goodness-of-Fit Test Results ---")
print(f"Observed Frequencies: {observed_frequencies}")
print(f"Expected Frequencies: {expected_frequencies_gof}")
print(f"Hypothesized Probabilities: {hypothesized_probabilities}")
print(f"\nChi-square Statistic: {chi2_statistic_gof:.4f}")
print(f"P-value: {p_value_gof:.4f}")
print(f"Degrees of Freedom: {dof_gof}")
print(f"Significance Level (alpha): {alpha_gof}")


if p_value_gof < alpha_gof:
    print(f"\nDecision: Reject the null hypothesis (H0).")
    print("Conclusion: The observed data significantly deviates from the hypothesized distribution (e.g., the die is likely not fair).")
else:
    print(f"\nDecision: Fail to reject the null hypothesis (H0).")
    print("Conclusion: There is not enough evidence to suggest that the observed data significantly deviates from the hypothesized distribution (e.g., the die appears consistent with being fair).")

# Visualize the Chi-square distribution and the test statistic location
visualize_chi_square(degrees_of_freedom=dof_gof)

# Add the calculated Chi-square statistic and critical value(s) to the visualization for GoF test
# Generate points for the distribution
x_gof_vis = np.linspace(0, chi2.ppf(0.999, dof_gof), 200)
pdf_gof_vis = chi2.pdf(x_gof_vis, dof_gof)

plt.figure(figsize=(10, 6))
plt.plot(x_gof_vis, pdf_gof_vis, label=f'Chi-square Distribution (df = {dof_gof})')

# Plot the calculated Chi-square statistic
plt.axvline(chi2_statistic_gof, color='red', linestyle='dashed', linewidth=2, label=f'Calculated œá¬≤: {chi2_statistic_gof:.4f}')

# For goodness-of-fit, the rejection region is typically in the right tail
# Find the critical value for a one-sided test (as goodness-of-fit is usually tested for deviation in any direction, which results in a larger Chi-square statistic)
critical_chi2_gof = chi2.ppf(1 - alpha_gof, dof_gof)

# Shade the rejection region (right tail)
x_fill_right_gof = x_gof_vis[x_gof_vis >= critical_chi2_gof]
pdf_fill_right_gof = pdf_gof_vis[x_gof_vis >= critical_chi2_gof]
plt.fill_between(x_fill_right_gof, pdf_fill_right_gof, color='orange', alpha=0.5, label=f'Rejection Region (Œ± = {alpha_gof})')

# Add the critical value line
plt.axvline(critical_chi2_gof, color='red', linestyle='dotted', linewidth=2, label=f'Critical œá¬≤: {critical_chi2_gof:.4f}')


plt.title('Chi-square Goodness-of-Fit Test')
plt.xlabel('Chi-square Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

print("\nVisual Interpretation of Goodness-of-Fit Test:")
print(f"- The curve shows the Chi-square distribution with {dof_gof} degrees of freedom.")
print(f"- The dashed red line indicates the location of the calculated Chi-square test statistic ({chi2_statistic_gof:.4f}).")
print(f"- The orange shaded region represents the rejection region in the right tail at alpha = {alpha_gof}.")
print(f"- The dotted red line marks the critical Chi-square value ({critical_chi2_gof:.4f}).")
if chi2_statistic_gof > critical_chi2_gof:
    print(f"- Since the calculated Chi-square statistic ({chi2_statistic_gof:.4f}) falls within the rejection region, we reject the null hypothesis.")
    print("  This visually supports the conclusion that the observed data deviates from the expected distribution.")
else:
    print(f"- Since the calculated Chi-square statistic ({chi2_statistic_gof:.4f}) falls outside the rejection region, we fail to reject the null hypothesis.")
    print("  This visually supports the conclusion that the observed data is consistent with the expected distribution.")

print("\nExplanation of Chi-square Goodness-of-Fit Test Steps:")
print("1. State the Hypotheses:")
print("   - Null Hypothesis (H0): The observed frequencies match the expected frequencies from the hypothesized distribution.")
print("   - Alternative Hypothesis (H1): The observed frequencies do not match the expected frequencies (the distribution is different).")
print("2. Set the Significance Level (alpha):")
print("   - Alpha (Œ±) is the probability of making a Type I error (rejecting H0 when it is true). A common value is 0.05 (5%).")
print("3. Calculate Expected Frequencies:")
print("   - Based on the total number of observations and the hypothesized probabilities for each category.")
print("4. Calculate the Chi-square Test Statistic:")
print("   - Formula: œá¬≤ = Œ£ [ (Observed - Expected)¬≤ / Expected ] for each category.")
print("   - This statistic measures the discrepancy between observed and expected frequencies.")
print("5. Determine Degrees of Freedom:")
print("   - Degrees of freedom (df) = Number of categories - 1.")
print("6. Find the P-value:")
print("   - The P-value is the probability of obtaining a Chi-square statistic as large as, or larger than, the calculated one, assuming the null hypothesis is true, based on the Chi-square distribution with the calculated degrees of freedom.")
print("7. Make a Decision:")
print("   - Compare the P-value to the alpha level.")
print("   - If P-value < alpha: Reject H0. There is a significant difference between observed and expected frequencies.")
print("   - If P-value ‚â• alpha: Fail to reject H0. There is not enough evidence to say the distributions are different.")
```
